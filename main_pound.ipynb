# Check completeness of transcription data

```python
from lib import *

gatherer = Gatherer()
with time_perf("Check transcriptions completeness"):
    df_check_transcriptions = gatherer.check_transcriptions_complete()
```

```python
df_check_transcriptions.head()
```

```python
bad_con = df_check_transcriptions['conversation_id'].tolist()
for entry in gatherer._iterate(filter=bad_con):
    zipper = zip(['audiophile', 'backbiter', 'cliffhanger'], gatherer._get_transcription(entry))
    for t, df in zipper:
        bad_info = df_check_transcriptions[
            (df_check_transcriptions['conversation_id'] == entry.name) & 
            (df_check_transcriptions['type'] == t)
        ]
        if bad_info.empty:
            continue
        row = bad_info['row'].values[0]
        print(f"--- {entry.name} {t} ({row}) ---")
        display(df.iloc[row])
        print()
```

```python
df_check_transcriptions.info()
```

```python
filter = df_check_transcriptions['conversation_id'].tolist()
for entry in gatherer._iterate(filter=filter):
    print(entry.name)
```

```python
for entry in gatherer._iterate(filter=filter):
    zipper = zip(['audiophile', 'backbiter', 'cliffhanger'], gatherer._get_transcription(entry))
    for t, df in zipper:
        if df is None:
            print(f"{entry.name} missing {t} transcription")
            continue
        
        print(f"{entry.name} {t} transcription shape: {df.shape}")

        # print rows with missing values
        missing = df[df.isnull().any(axis=1)]
        if not missing.empty:
            print(f"Missing values in {entry.name} {t} transcription:")
            print(missing)
```
